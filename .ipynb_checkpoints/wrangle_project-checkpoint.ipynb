#### Programmatic Download

```python
# import the requests and os library 
import requests
import os
```

```python
# programmatically download the image prediction tsv file
url = 'https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv'
response = requests.get(url)
```

```python
# save the contentent of the reponse to a .tsv file
with open('image_predictions.tsv', mode='wb') as file:
    file.write(response.content)
```

```python
# read the tsv file to a dataframe
predictions = pd.read_csv('image_predictions.tsv', sep='\t')
```

#### Querying the Twitter API for retweet and like count

```python
# import the access library for Twitter API
import tweepy 
```

```python
# define the keys and tokens needed
consumer_key = ''
consumer_key_secret =''

access_token = ''
access_token_secret = ''
```

```python
# get authorisation using the customer keys and secret
auth = tweepy.OAuthHandler(consumer_key, consumer_key_secret)

# get access using the access tokens
auth.set_access_token(access_token, access_token_secret)
```

```python
# call the API
api = tweepy.API(auth, wait_on_rate_limit=True)
```

<a id='step2'></a>

### Action for Querying the API



```python
# create list for tweet_id from the acrchive dataframe
ids = list(archive.tweet_id)
```

```python
import json
```

```python
# create a txt txt file and run the tweet ids through a loop.
# store each tweet status content as JSON format line by line in the txt file.
count = 0
with open('tweet_json.txt', 'w') as file:
    for each_id in ids:
        count += 1
        try:
            status = api.get_status(each_id, tweet_mode='extended')
            json.dump(status._json, file)
            file.write('\n')
            print(str(count), ': successful')
        
        except Exception as e:
            print('tweet not found: ', e)
            pass
```

```python
# read the json file line by line into a dataframe
frame_list = []
with open('tweet_json.txt') as open_file:
    for line in open_file.readlines():
        tweet_id = json.loads(line)['id']
        retweet_count = json.loads(line)['retweet_count']
        favorite_count = json.loads(line)['favorite_count']
        language = json.loads(line)['lang']
        frame_list.append([int(tweet_id), retweet_count, favorite_count, language])

columns = ['tweet_id', 'retweet_count', 'favorite_count', 'language']
counts_df = pd.DataFrame(data = frame_list, columns = columns)
```

#### Insight 2: 
Were the tweets made everday of the week or just select days? Does any particular weekday have substantially more or less tweets than others?

##### Process
- First, I'll create a column for the days of the week, then group by each day and obtain the count of tweets on those days.
- The dataframe used was the  **Master dataset: `ratings_data`**. It contains the merged `tweets_info` dataframe got after cleaning the `twitter _archive_enhanced.csv` file, and the `predictions_clean` dataframe from cleaning the `predictions` provded from the neural network. Also the retweet count and favorite count, lthough not used for this insight.
- The **Master dataset** provides data for all tweets that were predictably confirmed to be dogs by the neural network

```python
# import datetime library
# create column for the weekday info
from datetime import datetime as dt
weekday = ratings_data.timestamp.dt.strftime('%A')

# add the week day column to the ratings_data dataframe
ratings_data['weekday'] = weekday
```

```python
# groupby the weekday and obtain the count for each
tweet_days = ratings_data.groupby('weekday')['tweet_id'].count()
tweet_days
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tweet_days.sort_values().plot(kind='barh')
plt.title('Tweeet Count by Weekday', fontsize=18)
plt.xlabel('count', fontsize=13)
plt.ylabel('weekday', fontsize=13);
```

##### Deduction:

The user tweeted on every day of the week. Similar the tweet distribution is even through the week. However, **Monday** seemed to have slightly more tweets than any other day in the dataset.

#### Insight 3: 
What form of engagements do the tweets receive? Do followers like more tweets than they share?

##### Process
I will be using the **Master dataset: `ratings_data`** as it contains a the columns for *retweet* and *favorite* count.
It also confirms that only tweets with positive predictions of being dogs are analysed. 

```python
retweets = ratings_data.retweet_count.sum()
favorites = ratings_data.favorite_count.sum()

print('Total retweets for all tweets was: ', retweets, '\nTotal Favorites (likes) for all tweets was: ', favorites)
```

```python
plt.bar('retweets', retweets)
plt.bar('favorites', favorites)
plt.title('Total number of Retweets and Favorites', fontsize= 15)
plt.ylabel('count (in tens of millions)', fontsize= 13);
```

##### Deduction:
Significantly more viewers prefered to just like the tweets than to retweet.

#### Insight 4:
Did the ratings have any effect on the engagements by viewers?

##### Process: 

I will be using the **Master dataset: `ratings_data`** to compare the rating score and the amount of user engagement typically gotten for that score.

```python
rating_engagement = ratings_data.groupby('rating_numerator')[['retweet_count', 'favorite_count']].sum()
rating_engagement
```

```python
# make a plot to visualise the differences
rating_engagement.plot(kind='bar',figsize=(8,8))
plt.title('Count of retweets and favorites by dog rating', fontsize = 15)
plt.ylabel('count (in millions)', fontsize = 13)
plt.xlabel('dog rating', fontsize = 13);
```

##### Deduction:

From the chart and analysis above, more viewers engage the tweets with ratings between 10 and 14.
